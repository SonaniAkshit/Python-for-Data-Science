# âœ¨ Python Special Methods (`__dunder__` methods)

Special methods in Python start and end with double underscores (`__`) and allow **custom behavior** for classes â€” like operator overloading, object initialization, representation, and more.

---

### ğŸ§± 1. `__init__()` â€“ Constructor

Called when an object is created.

```python
class Person:
    def __init__(self, name):
        self.name = name
````

ğŸ› ï¸ Initializes the object with properties

---

### ğŸ“¦ 2. `__str__()` â€“ String Representation

Defines what `str(obj)` or `print(obj)` returns.

```python
def __str__(self):
    return f"Person({self.name})"
```

ğŸ“„ Used for **user-friendly output**

---

### ğŸ§  3. `__repr__()` â€“ Official Representation

Defines what `repr(obj)` returns (used in debugging, shells).

```python
def __repr__(self):
    return f"Person(name={self.name})"
```

ğŸ” More **detailed and developer-oriented**

---

### â• 4. `__add__(self, other)` â€“ `+` Operator

Used for **operator overloading**.

```python
def __add__(self, other):
    return self.value + other.value
```

ğŸ”— You can also overload `__sub__`, `__mul__`, etc.

---

### ğŸŸ° 5. `__eq__(self, other)` â€“ Equality `==`

Defines custom behavior for the `==` operator.

```python
def __eq__(self, other):
    return self.name == other.name
```

âœ… Used to compare two objects

---

### ğŸ”¢ 6. `__len__(self)` â€“ `len(obj)`

Defines behavior of `len()` on your object.

```python
def __len__(self):
    return len(self.data)
```

ğŸ“ Use this when your class acts like a container

---

### ğŸ§± 7. `__getitem__(self, index)` â€“ Indexing

Enables bracket indexing: `obj[index]`

```python
def __getitem__(self, index):
    return self.data[index]
```

ğŸ“¦ Make your object behave like a list or dict

---

### âœï¸ 8. `__setitem__(self, index, value)` â€“ Item Assignment

```python
def __setitem__(self, index, value):
    self.data[index] = value
```

ğŸ”§ Supports `obj[index] = value`

---

### ğŸ—‘ï¸ 9. `__delitem__(self, index)` â€“ Deletion

```python
def __delitem__(self, index):
    del self.data[index]
```

ğŸ§¹ Enables `del obj[index]`

---

### ğŸ” 10. `__iter__()` + `__next__()` â€“ Iterable Support

```python
def __iter__(self):
    return self

def __next__(self):
    # your logic
    raise StopIteration
```

ğŸ”„ Make your object work in a `for` loop

---

### ğŸ”’ 11. `__contains__(self, item)` â€“ `in` keyword

```python
def __contains__(self, item):
    return item in self.data
```

ğŸ” Enables `item in obj`

---

### ğŸ”¢ 12. `__call__()` â€“ Callable Object

```python
def __call__(self, x):
    return x * 2
```

ğŸ“ Lets you "call" the object like a function: `obj(3)`

---

### ğŸ§¹ 13. `__del__()` â€“ Destructor

Called when object is about to be destroyed (use cautiously).

```python
def __del__(self):
    print("Object deleted")
```

ğŸ’€ Use only when cleanup is essential

---

## ğŸ§  Bonus: All Operator Overload Methods

| Operator | Method           |
| -------- | ---------------- |
| `+`      | `__add__()`      |
| `-`      | `__sub__()`      |
| `*`      | `__mul__()`      |
| `/`      | `__truediv__()`  |
| `//`     | `__floordiv__()` |
| `%`      | `__mod__()`      |
| `**`     | `__pow__()`      |
| `==`     | `__eq__()`       |
| `!=`     | `__ne__()`       |
| `<`      | `__lt__()`       |
| `<=`     | `__le__()`       |
| `>`      | `__gt__()`       |
| `>=`     | `__ge__()`       |

---

### ğŸ’¡ Tip:

* Special methods help integrate your class **seamlessly with Python syntax**
* Great for building **custom data structures, math classes, iterators, and more**
