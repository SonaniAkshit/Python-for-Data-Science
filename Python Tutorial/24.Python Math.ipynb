# ‚ûó Python `math` Module

The `math` module provides **mathematical functions** like square roots, trigonometry, logarithms, constants, etc.

---

### üîπ Import the Module

```python
import math
```

---

### 1Ô∏è‚É£ Common Math Functions

```python
print(math.sqrt(16))     # ‚úÖ Square root ‚Üí 4.0
print(math.pow(2, 3))    # ‚úÖ Power ‚Üí 8.0
print(math.factorial(5))# ‚úÖ Factorial ‚Üí 120
print(math.fabs(-10))   # ‚úÖ Absolute value ‚Üí 10.0
```

---

### 2Ô∏è‚É£ Rounding Functions

```python
print(math.floor(3.9))   # ‚¨áÔ∏è Floor ‚Üí 3
print(math.ceil(3.2))    # ‚¨ÜÔ∏è Ceil ‚Üí 4
print(math.trunc(5.99))  # ‚úÇÔ∏è Truncate ‚Üí 5
```

---

### 3Ô∏è‚É£ Trigonometric Functions

```python
print(math.sin(math.pi/2))  # sin(90¬∞) ‚Üí 1.0
print(math.cos(0))          # cos(0¬∞) ‚Üí 1.0
print(math.tan(math.pi/4))  # tan(45¬∞) ‚Üí ~1.0
```

---

### 4Ô∏è‚É£ Logarithmic & Exponential Functions

```python
print(math.log(100, 10))   # log base 10 ‚Üí 2.0
print(math.log2(8))        # log base 2 ‚Üí 3.0
print(math.exp(2))         # e^2 ‚Üí ~7.39
```

---

### 5Ô∏è‚É£ Constants

```python
print(math.pi)      # œÄ = 3.141592653...
print(math.e)       # e = 2.718281828...
print(math.tau)     # œÑ = 6.283185307...
```

---

### üßæ Summary Table

| Function            | Description                         |
| ------------------- | ----------------------------------- |
| `math.sqrt(x)`      | Square root of `x`                  |
| `math.pow(x, y)`    | `x` to the power `y`                |
| `math.factorial(x)` | Factorial of `x`                    |
| `math.floor(x)`     | Round down                          |
| `math.ceil(x)`      | Round up                            |
| `math.trunc(x)`     | Truncate to integer                 |
| `math.log(x, base)` | Logarithm base `base` (default `e`) |
| `math.sin(x)`       | Sine of `x` radians                 |
| `math.pi`, `math.e` | Math constants                      |

---
